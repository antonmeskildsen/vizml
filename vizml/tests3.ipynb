```python
from graph import Session
from nodes import Variable, Constant, Input, Output
from IPython.display import display
import numpy as np
```

```python
X = Input('X')
V = Variable(2, 'V')

add_node = V+Constant(0.5)
out = Output(X*(add_node))

ctx = {
    'X': 3
}

out
```

```python
for node in out.graph.topological():
    node.compute(ctx)
    print(f'computed:  {node.__str__():10s} -> {node.value}')
```
